# IF ELSE and ELIF

```python
mark = 90
```

```python
if mark >= 91 :
    print("Grade -> S")
elif mark >= 81 and mark <91:
    print("Grade-> A")
elif mark >= 71 and mark <81:
    print("Grade-> B")
elif mark >= 61 and mark <71:
    print("Grade-> C")
elif mark >= 57 and mark <61:
    print("Grade-> D")
elif mark >= 50 and mark <57:
    print("Grade-> E")
else:
    print("Grade-> U")
```

# For Loop

```python
for x in range(1,1000):
    if x > 1:        
        for y in range(2, x):
            if (x % y == 0) :
                break
        else:
            print(x)
```

```python
for a in range(1,11):
    print(a,"table")
    for b in range(1,11):
        print(b,"X",a,"=",b*a)
    print("")
```

```python
primeinput = int(input("Enter Any Number: "))

Number = 1

print("Prime numbers between", 1, "and", primeinput, "are:")

while(Number <= primeinput):
    count = 0
    i = 2

    while(i <= Number//2):
        if(Number % i == 0):
            count = count + 1
            break
        i = i + 1

    if (count == 0 and Number != 1):
        print(" %d" %Number, end = '  ')
    Number = Number  + 1
```
